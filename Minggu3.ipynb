**Langkah 6:** Regresi Linier

```python
# Membuat variabel bebas X dan Y, contoh pengambilan dari analisis korelasi sebelumnya
X = data['Length of Membership']
y = data['Yearly Amount Spent']

print(X.head())
print(y.head())
```

**Analisis**
- X: Variabel ini mewakili data dari kolom 'Length of Membership' yang merupakan durasi keanggotaan pelanggan.
- y: Variabel ini mewakili data dari kolom 'Yearly Amount Spent' yang merupakan jumlah uang yang dibelanjakan pelanggan per tahun.
- `print(X.head())`: Menampilkan 5 baris pertama dari variabel X untuk memberikan gambaran tentang data yang terkandung dalam kolom 'Length of Membership'.
- `print(y.head())`: Menampilkan 5 baris pertama dari variabel y untuk memberikan gambaran tentang data yang terkandung dalam kolom 'Yearly Amount Spent'.

- Pisahkan variabel bebas (X) dan variabel target (y).

```python
# Pembagian data latih dan data uji dengan proporsi 7:3
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)

print("Ukuran data latih:", X_train.shape[0])
print("Ukuran data uji:", X_test.shape[0])
print("Beberapa baris data latih:")
print(X_train.head())
print("Beberapa baris data uji:")
print(X_test.head())
```

**Analisis**
- `train_test_split` adalah fungsi dari pustaka scikit-learn yang digunakan untuk membagi dataset menjadi data latih dan data uji.
- X: Fitur atau variabel bebas yang digunakan untuk melatih model.
- y: Target atau variabel yang ingin diprediksi.
- `train_size=0.7`: Menentukan bahwa 70% dari data akan digunakan sebagai data latih.
- `test_size=0.3`: Menentukan bahwa 30% dari data akan digunakan sebagai data uji.
- `random_state=100`: Menetapkan nilai seed untuk memastikan pembagian data yang konsisten setiap kali kode dijalankan (reproducibility).
- Fungsi `train_test_split` akan mengembalikan empat variabel:
- `X_train`: Data latih untuk fitur.
- `X_test`: Data uji untuk fitur.
- `y_train`: Data latih untuk target.
- `y_test`: Data uji untuk target.

- Bagi data menjadi data latih (70%) dan data uji (30%) menggunakan train_test_split.
- Lakukan training model regresi linier menggunakan library StatsModels. Tambahkan konstanta (intercept) ke variabel bebas.

```python
# Training model
import statsmodels.api as sm

X_train_sm = sm.add_constant(X_train)
lr = sm.OLS(y_train, X_train_sm).fit()

print(lr.summary())
```

**Analisis** Kode ini digunakan untuk melatih model regresi linier menggunakan pustaka statsmodels di Python.

- `statsmodels` adalah pustaka yang digunakan untuk berbagai jenis analisis statistik, termasuk regresi linier.
- `sm.add_constant(X_train)`: Menambahkan kolom konstanta (bias) ke data fitur X_train. Ini penting karena model regresi linier sering memerlukan intercept (konstanta) untuk menghitung regresi yang tepat.
- `sm.OLS(y_train, X_train_sm)`: Membuat objek model regresi linier ordinary least squares (OLS) menggunakan data latih X_train_sm dan target y_train.
- `.fit()`: Melatih model dengan data yang diberikan dan mengembalikan objek model yang sudah dilatih.
- `lr.summary()`: Menampilkan ringkasan hasil dari model regresi linier yang sudah dilatih



- Visualisasikan garis regresi pada data latih.

```python
# Visualisasi garis regresi
plt.scatter(X_train, y_train)
plt.plot(X_train, 265.2483 + 66.3015*X_train, 'r')
plt.show()
```

**Analisis**
Kode ini digunakan untuk memvisualisasikan hasil regresi linier dengan menggambar grafik scatter plot dari data latih dan menambahkan garis regresi pada grafik tersebut.

- `plt.scatter(X_train, y_train)`: Menggambar scatter plot (grafik titik) yang menunjukkan hubungan antara variabel bebas X_train dan variabel tergantung y_train
- `plt.plot(...)`: Menggambar garis pada grafik.
- `265.2483 + 66.3015*X_train`: persamaan garis regresi linier yang digunakan untuk menggambar garis pada scatter plot. Angka 265.2483 adalah intercept (konstanta), dan 66.3015 adalah koefisien regresi (gradien) yang menunjukkan kemiringan garis.
- `'r'`: Menentukan warna garis, dalam hal ini 'r' berarti merah.

**Langkah 7** Analisis Residual

```python
# Prediksi nilai y_value dari data x yang telah dilatih
y_train_pred = lr.predict(X_train_sm)

# Menghitung residual
res = (y_train - y_train_pred)

import matplotlib.pyplot as plt

plt.scatter(X_train, res)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Length of Membership')
plt.ylabel('Residuals')
plt.title('Plot Residuals vs. Fitted Values')
plt.show()
```

**Analisis** Kode ini digunakan untuk membuat prediksi dengan model regresi linier dan menghitung residuals (selisih antara nilai sebenarnya dan nilai yang diprediksi oleh model)
- `lr.predict(X_train_sm)`: Menggunakan model regresi linier yang telah dilatih (lr) untuk memprediksi nilai y berdasarkan data fitur X_train_sm yang telah ditambahkan kolom konstanta.
- `y_train_pred`: Variabel ini menyimpan hasil prediksi untuk y dari data latih X_train. Ini akan berisi nilai-nilai yang diprediksi oleh model regresi linier untuk setiap nilai dalam X_train_sm.
- `y_train` - y_train_pred: Menghitung residuals, yaitu selisih antara nilai sebenarnya (y_train) dan nilai yang diprediksi oleh model (y_train_pred).
- `res`: Variabel ini menyimpan nilai residual untuk setiap data dalam X_train. Residuals menunjukkan seberapa jauh prediksi model dari nilai sebenarnya.
- `plt.scatter(X_train, res)`: Menggambar scatter plot (grafik titik) yang menunjukkan hubungan antara nilai fitur (X_train) dan residuals (res)
- `plt.axhline(y=0, color='r', linestyle='--')`: Menambahkan garis horizontal pada sumbu y di nilai 0.

- Lakukan prediksi nilai y dari data latih dan hitung residual (selisih antara nilai sebenarnya dan nilai prediksi).

```python
# Histogram residual
fig = plt.figure()
sns.distplot(res, bins=15)
plt.title('Error Terms', fontsize=15)
plt.xlabel('y_train - y_train_pred', fontsize=15)
plt.show()

# Scatter plot residual
plt.scatter(X_train, res)
plt.show()
```

**Analisis** Kode ini digunakan untuk menganalisis residuals dari model regresi linier dengan dua jenis visualisasi

- `fig = plt.figure()`: Membuat sebuah objek figure baru untuk visualisasi. Ini berguna jika kamu ingin membuat beberapa grafik dalam satu skrip.
- `sns.distplot(res, bins=15)`: Menggunakan seaborn untuk membuat histogram dari residuals (res).
- `bins=15`: Menentukan jumlah bin (kelompok) pada histogram. Ini membagi rentang data menjadi 15 kelompok.
- `plt.title('Error Terms', fontsize=15)`: Menambahkan judul pada histogram untuk memberikan konteks tentang apa yang ditampilkan.
- `plt.xlabel('y_train - y_train_pred', fontsize=15)`:
Menambahkan label pada sumbu x yang menunjukkan residuals (selisih antara nilai yang sebenarnya dan nilai yang diprediksi).
- `plt.scatter(X_train, res)`: Menggambar scatter plot dari residuals (res) terhadap nilai fitur (X_train).

- Visualisasikan residual dalam bentuk histogram dan scatter plot untuk mengevaluasi distribusi dan pola error.

**Langkah 8:** Prediksi pada Data Uji dan Evaluasi Model

```python
# Prediksi pada data uji
X_test_sm = sm.add_constant(X_test)
y_test_pred = lr.predict(X_test_sm)

print(y_test_pred[:10])
```

**Analisis** Kode ini digunakan untuk membuat prediksi pada data uji (testing data) menggunakan model regresi linier yang telah dilatih sebelumnya.

- `sm.add_constant(X_test)`: Menambahkan kolom konstanta (bias) ke data fitur X_test yang akan digunakan untuk prediksi.
- `lr.predict(X_test_sm)`: Menggunakan model regresi linier (lr) untuk memprediksi nilai target (y) berdasarkan data uji (X_test_sm).
- `y_test_pred`: Variabel ini menyimpan hasil prediksi untuk data uji.
- `print(y_test_pred[:10])`: Menampilkan 10 prediksi pertama


- Lakukan prediksi pada data uji.

```python
# Hitung nilai R-squared
from sklearn.metrics import r2_score

r_squared = r2_score(y_test, y_test_pred)

print(f'Nilai R-squared: {r_squared:.2f}')
```

**Analisis** Kode ini digunakan untuk menghitung nilai R-squared (koefisien determinasi) dari model regresi linier menggunakan data uji.

- `r2_score` adalah fungsi dari pustaka scikit-learn yang digunakan untuk menghitung nilai R-squared.
- `r2_score(y_test, y_test_pred)`: Fungsi ini menghitung R-squared dengan membandingkan nilai sebenarnya dari data uji (y_test) dengan nilai yang diprediksi oleh model (y_test_pred).
- `r_squared`: Variabel ini menyimpan nilai R-squared yang dihitung.

- Hitung nilai R-squared untuk mengukur kinerja model pada data uji.

```python
# Mengimpor dataset (Pastikan Anda memiliki file CSV 'Posisi_gaji.csv' dalam direktori yang sama)
dataset = pd.read_csv('Posisi_gaji.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values  # Ubah menjadi satu kolom saja

print('Data X: ', X)
print('Data y: ', y)

dataset.info()
```

**3. Feature Scaling:**

Menggunakan StandardScaler untuk melakukan penskalaan fitur X dan target y. Ini diperlukan karena SVM sangat sensitif terhadap skala data.

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X = sc_X.fit_transform(X.reshape(-1, 1))
y = sc_y.fit_transform(y.reshape(-1, 1))

print('Data X setelah di featuring scaling: ', X)
print('Data Y setelah di featuring scaling: ', y)
```

**Analisis** Feature scaling adalah teknik untuk merubah skala data sehingga menjadi lebih mudah digunakan dalam analisis atau pelatihan model machine learning.

- `from sklearn.preprocessing import StandardScaler` ini mengimpor StandardScaler dari sklearn.preprocessing, yang digunakan untuk menstandardisasi data.
- `sc_X = StandardScaler()` inisialisasi StandardScaler untuk fitur X dan satu untuk target y
- `X.reshape(-1, 1) dan y.reshape(-1, 1)` mengubah bentuk data menjadi array dua dimensi dengan satu kolom. Ini diperlukan karena StandardScaler bekerja dengan array dua dimensi.
- `fit_transform `digunakan untuk menghitung rata-rata dan deviasi standar dari data, kemudian mentransformasikan data ke skala standar (mean = 0, deviasi standar = 1).


**4. Fitting SVR ke Dataset:**

Lakukan pembuatan model SVR dengan kernel RBF (Radial Basis Function) dan melatihnya dengan data yang telah di-scaled.

```python
# Fitting SVR ke dataset
from sklearn.svm import SVR
regressor = SVR(kernel='rbf')
regressor.fit(X, y)
```

**Analisis** Kode ini digunakan untuk melatih model Support Vector Regression (SVR) pada dataset yang telah diskalakan.
- `from sklearn.svm import SVR` mengimpor SVR dari sklearn.svm, yang adalah algoritma untuk regresi menggunakan Support Vector Machine (SVM)
- `SVR` adalah kelas yang digunakan untuk regresi dengan metode Support Vector Machine.
- `kernel='rbf'` menunjukkan bahwa kernel yang digunakan adalah Radial Basis Function (RBF). Kernel ini memungkinkan model menangani hubungan non-linear antara fitur dan target.
- `fit` adalah metode yang digunakan untuk melatih model dengan data.
- `X` adalah fitur atau data input, dan y adalah target atau nilai yang ingin diprediksi.


**5. Visualisasi Hasil SVR:**

langkah selanjutkan, lakukan visualisasi Menggunakan grafik untuk memvisualisasikan hasil prediksi model SVR. Ini mencakup plotting data asli (titik-titik merah) dan kurva hasil prediksi (garis biru) untuk tingkat posisi yang bervariasi.

```python
# Visualisasi hasil SVR (resolusi tinggi dan kurva yang lebih halus)
X_grid = np.arange(min(X), max(X), 0.01).reshape(-1, 1)
plt.scatter(X, y, color='red')
plt.plot(X_grid, regressor.predict(X_grid), color='blue')
plt.title('Jujur atau tidak (SVR)')
plt.xlabel('Tingkat posisi')
plt.ylabel('Gaji')
plt.show()
```

**Analisis** Kode ini digunakan untuk memvisualisasikan hasil dari model SVR (Support Vector Regression) dengan kurva yang lebih halus dan resolusi tinggi.

- `np.arange(min(X), max(X), 0.01)` membuat array dengan rentang nilai dari yang terkecil hingga yang terbesar di X, dengan interval 0.01. Ini menciptakan data grid yang lebih halus untuk plot.
- `reshape(-1, 1)` mengubah array menjadi bentuk dua dimensi dengan satu kolom, sesuai format input yang diharapkan oleh regressor.predict().
- `plt.scatter(X, y, color='red')` membuat plot titik-titik data asli (X dan y) dengan warna merah.
- `regressor.predict(X_grid)` memprediksi nilai y untuk setiap nilai di X_grid.
- `plt.plot(X_grid, regressor.predict(X_grid), color='blue')` menggambar kurva hasil prediksi model SVR dengan warna biru. Ini menunjukkan bagaimana model memprediksi data pada interval yang lebih halus.

**6. Prediksi Hasil:**

- Membuat array 2D yang berisi tingkat posisi yang akan diprediksi. Dalam contoh ini, tingkat posisi 6.5.

- Menskalakan fitur prediksi menggunakan sc_X.transform.

- Melakukan prediksi menggunakan model SVR yang telah dilatih.

- Mengembalikan hasil prediksi ke dalam skala aslinya menggunakan sc_y.inverse_transform.

```python
# Prediksi hasil
# Buat array 2D yang berisi tingkat posisi yang akan diprediksi
tingkat_posisi_prediksi = np.array([[6.5]])
# Penskalaan fitur untuk data yang akan diprediksi
tingkat_posisi_prediksi = sc_X.transform(tingkat_posisi_prediksi)
# Melakukan prediksi menggunakan model SVR
gaji_prediksi = regressor.predict(tingkat_posisi_prediksi)
# Kembalikan hasil prediksi ke skala aslinya
gaji_prediksi = sc_y.inverse_transform(gaji_prediksi.reshape(-1, 1))

print(f'Prediksi gaji untuk tingkat posisi 6.5 adalah: {gaji_prediksi[0][0]:.2f}')
```

**Analisis** Kode ini digunakan untuk membuat prediksi dengan model Support Vector Regression (SVR) pada data baru dan mengembalikan hasil prediksi ke skala aslinya.

- `tingkat_posisi_prediksi `adalah array dua dimensi yang berisi nilai tingkat posisi yang ingin diprediksi. Dalam hal ini, nilai tersebut adalah 6.5.
- `sc_X.transform` digunakan untuk menstandardisasi data prediksi sesuai dengan skala yang telah diterapkan pada data pelatihan (X).
- `regressor.predict` digunakan untuk memprediksi nilai gaji berdasarkan tingkat_posisi_prediksi yang telah diskalakan.
- `sc_y.inverse_transform `mengembalikan hasil prediksi ke skala aslinya, yaitu mengembalikan data ke skala yang sesuai dengan data asli sebelum dilakukan penskalaan.
- `reshape(-1, 1)` mengubah bentuk array gaji_prediksi menjadi dua dimensi dengan satu kolom, sesuai format yang dibutuhkan oleh inverse_transform.

**Evaluasi Model SVR**

Langkah terakhir adalah melakukan evaluasi model meliputi MAE, MSE dan R-squared

```python
# Evaluasi model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_actual = y
y_pred = regressor.predict(X)

# Menghitung MAE
mae = mean_absolute_error(y_actual, y_pred)

# Menghitung MSE
mse = mean_squared_error(y_actual, y_pred)

# Menghitung RMSE
rmse = np.sqrt(mse)

# Menghitung R-squared
r2 = r2_score(y_actual, y_pred)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R-squared:", r2)
```

**Analisis**

- `mean_absolute_error, mean_squared_error, dan r2_score `adalah fungsi dari sklearn.metrics yang digunakan untuk menghitung metrik evaluasi model
- `y_actual `adalah nilai asli dari target (gaji) dari dataset.
- `y_pred` adalah nilai prediksi yang dihasilkan oleh model SVR untuk dataset tersebut.
- `MAE` mengukur rata-rata perbedaan absolut antara nilai aktual dan nilai prediksi. Semakin kecil nilai MAE, semakin baik model.
- `MSE` mengukur rata-rata kuadrat dari perbedaan antara nilai aktual dan nilai prediksi. MSE memberikan penalti lebih besar pada kesalahan yang lebih besar.
- `RMSE` adalah akar kuadrat dari MSE. Ini memberikan ukuran kesalahan dalam satuan yang sama dengan data asli dan sering kali lebih mudah diinterpretasikan.
- `R-squared` mengukur seberapa baik model menjelaskan variabilitas dalam data. Nilai R berkisar antara 0 hingga 1, dengan nilai yang lebih tinggi menunjukkan model yang lebih baik.
